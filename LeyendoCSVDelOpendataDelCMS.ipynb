Algunos de los archivos para educación del CMS con datos abiertos están en un formato CSV. Este formato es un formato de texto sencillo en donde cada evento se guarda en columnas y cada columna correspende a una variable cinemática reconstruida con el detector. Se puede encontrar un ejemplo de esta liga
https://opendata.cern.ch/record/5208.
En partícular estos datos presentan las variables cinemáticas de dos muones producto final de la colisión protón-protón. Las variables se pueden determinar usando los cuadrimomentos de las partículas (https://es.wikipedia.org/wiki/Cuadrimomento)

```python
#panda nos permite manipular el archivo csv https://pandas.pydata.org
import plotly.graph_objects as go

import pandas as pd
#numpy nos permite hacer operaciones usando vectores https://numpy.org
import numpy as np
#plotly nos permite crear graficas interactivas https://plotly.com/python/
import plotly.express as px
%matplotlib inline
#usamos la liga que está directamente en la página. 
particles = pd.read_csv('https://opendata.cern.ch/record/5208/files/Zmumu.csv',delimiter=',')
particles.head(50)
```

Los nombres de las columnas se refieren a las variables que describen a llos dos muones. Corrida (Etiqueta del haz), Número de evento (colisión), Momento Transverso del primer muon, Eta del primer muón, Phi del primer muón, Carga del primer muón, distancia al vertice, lo mismo para el segundo muon y finalmente la variable de aislamiento que se relaciona con la energía de las trazas que rodean al muon. 

```python
#función para reconstruir la masa invariante de las dos partículas. Usamos la definición encontrada https://en.wikipedia.org/wiki/Invariant_mass
def minvariant(pt1,pt2,eta1,eta2,phi1,phi2):
  return np.sqrt(2*pt1*pt2*(np.cosh(eta2-eta1)-np.cos(phi1-phi2)))
#de acuerdo a la formula se están reconstruyendo particulas relativistas unicamente (sin masa o con mucha energía)
 
minv=minvariant(particles['pt1'],particles['pt2'],particles['eta1'],particles['eta2'],particles['phi1'],particles['phi2'])

#Se crea y se grafica el histograma. 
fig_2=px.histogram(minv,nbins=200)
fig_2.update_layout(title=r"Masa Invariante M", xaxis_title=r"GeV",yaxis_title="Eventos")
fig_2.show()

fig_3=px.histogram(particles['phi1'])
fig_3.show()

#Se crea y se grafica el histograma. 
fig_5= px.scatter(x=particles['eta1'],y=particles['pt1'])
fig_5.update_layout(xaxis_title=r"GeV",yaxis_title="Eventos",yaxis_range=[0,40])
fig_5.show()

fig_2=px.histogram(particles['Q1']+particles['Q2']) #notemos que corresponden a un Z0 por lo que la mayoria de los eventos debe tener carga 0, en efecto
fig_2.show()
#teniendo el momento total podemos hacer un histograma de el.
```

```python
#Ahora vamos a obtener el momento transversal total de todo el evento de la siguiente forma:
def ps(pt1,pt2,phi1,phi2):
  return pt1*np.cos(phi1)+pt2*np.cos(phi2)

p_x=ps(particles['pt1'],particles['pt2'],particles['phi1'],particles['phi2'])

def py(pt1,pt2,phi1,phi2):
  return pt1*np.sin(phi1)+pt2*np.sin(phi2)

p_y=py(particles['pt1'],particles['pt2'],particles['phi1'],particles['phi2'])

p_tot=np.sqrt((p_x)**2+(p_y)**2)
print(p_tot.head(20))
figura=px.histogram(p_tot)
figura.show()



```

```python

datosfiltro0=particles [(particles.Q1+particles.Q2 == 0)]
print(len(datosfiltro0))
datosfiltro1=datosfiltro0[np.abs((datosfiltro0.dxy1<.02)) & np.abs((datosfiltro0.dxy2<.02))]
print(len(datosfiltro1))

minvf=minvariant(datosfiltro0['pt1'],datosfiltro0['pt2'],datosfiltro0['eta1'],datosfiltro0['eta2'],datosfiltro0['phi1'],datosfiltro0['phi2'])
fig = go.Figure()
fig.add_trace(go.Histogram(x=minvf, name='datos filtrados'))
fig.add_trace(go.Histogram(x=minv, name='datos sin filtro'))
fig.update_layout(barmode='stack')
fig.show()

fig2= go.Figure()
fig2.add_trace(go.Histogram(x=datosfiltro0['phi1'], name='datos filtrados'))
fig2.add_trace(go.Histogram(x=particles['phi1'], name='datos sin filtro'))
fig2.update_layout(barmode='stack')
fig2.show()

fig3= go.Figure()
fig3.add_trace(go.Histogram(x=datosfiltro0['eta1'], name='datos filtrados'))
fig3.add_trace(go.Histogram(x=particles['eta1'], name='datos sin filtro'))
fig3.update_layout(barmode='stack')
fig3.show()

p_xf=ps(datosfiltro0['pt1'],datosfiltro0['pt2'],datosfiltro0['phi1'],datosfiltro0['phi2'])
p_yf=py(datosfiltro0['pt1'],datosfiltro0['pt2'],datosfiltro0['phi1'],datosfiltro0['phi2'])
p_totf=np.sqrt((p_xf)**2+(p_yf)**2)
fig4= go.Figure()
fig4.add_trace(go.Histogram(x=p_totf, name='datos filtrados'))
fig4.add_trace(go.Histogram(x=p_tot, name='datos sin filtro'))
fig4.update_layout(barmode='stack')
fig4.show()
```
